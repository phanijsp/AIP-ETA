```python
import pandas as pd
import re

def clean_title(title):
    return re.sub(r'[^a-zA-Z0-9\s]', '', title)


df = pd.read_csv('titles.csv')


df['Title'] = df['Title'].apply(clean_title)

df.to_csv('cleaned_titles.csv', index=False)
```

```python
import pandas as pd
import requests
import time

def search_movies(csv_file_path, base_url):
    df = pd.read_csv(csv_file_path)

    for index, row in df.iloc[149:].iterrows():
        movie_title = row['Title']
        query_param = {'q': movie_title}
        
        try:
            response = requests.get(base_url, params=query_param)
            response.raise_for_status()  
            
            with open(f"results/{movie_title}.xml", 'wb') as file:
                file.write(response.content)
            
            print(f"Successfully saved {index}. {movie_title}.xml")
        
        except requests.exceptions.RequestException as e:
            print(f"Error searching for {movie_title}: {e}")
        
        time.sleep(10)


csv_file_path = 'cleaned_titles.csv'
base_url = 'http://127.0.0.1:9117/api/v2.0/indexers/all/results/torznab/api'
api_key = '7wgsc931bpo8ie1c656v1itc8ovqqgm2'

full_url = f"{base_url}?apikey={api_key}&t=search"

try:
    search_movies(csv_file_path, full_url)
except Exception as e:
    print(f"Error: {e}")
```
